### 1. Average salary based on job_title and employee_type

```python
temp_df=df_duplicated.groupby(["employment_type","job_title"])["salary_in_usd"].mean().sort_values(ascending=False).reset_index()
temp_df_new = pd.DataFrame({
    'employment_type':[],
    'job_title':[],
    'salary_in_usd' : []
})
for type in temp_df["employment_type"].unique():
    # print(temp_df[temp_df["employment_type"]==type].head())
     temp_df_new = pd.concat([temp_df[temp_df["employment_type"] == type].head(), temp_df_new])

temp_df_new
```

```python
temp_df = df_duplicated.groupby(['employment_type', 'job_title'])["salary_in_usd"].mean().sort_values(ascending = False).reset_index()

for type in temp_df["employment_type"].unique():
    sns.barplot(data  = temp_df[temp_df["employment_type"] == type].head()\
              , x = "job_title", y = "salary_in_usd", hue = "employment_type")
    plt.title(f"Top 5 average salaries based on job title for {type}" )
    plt.xticks(rotation = 45)
    plt.show()
```

CONCLUSION:   As we can see by our observation that more companies are realizing that they need a solid data infrastructure to add value to their business before hiring a Data Scientist or a Data ML Engineer.

### 2. Top 5 Average salary based on job title and exp level

```python
temp_df=df_duplicated.groupby(['experience_level','job_title'])["salary_in_usd"].mean()\
.sort_values(ascending=False).reset_index()
for exp in temp_df["experience_level"].unique():
    print(temp_df[temp_df["experience_level"]==exp].head())
```

```python
temp_df = df_duplicated.groupby(['experience_level', 'job_title'])["salary_in_usd"].mean().sort_values(ascending = False).reset_index()

for exp in temp_df["experience_level"].unique():
    sns.barplot(data  = temp_df[temp_df["experience_level"] == exp].head()\
              , x = "job_title", y = "salary_in_usd", hue = "experience_level")
    plt.title(f"Top 5 average salaries based on job title for {exp} level")
    plt.xticks(rotation = 45)
    plt.show()
```

 CONCLUSION:As we can see by our observation that Deep Learning engineer have more scope in the entry level(Experience), Data infrastructure Engineer in MI, Data Analytics Lead in SE and Principal Data Scientist in the EX.

## **3. Trending job_titles in india year wise**

```python
temp_df = df[["work_year", "job_title"]][df["company_location"]=="IN"].value_counts().reset_index(name = "jobs")

for year in temp_df["work_year"].unique():
    data = temp_df[temp_df["work_year"] == year].head()
    sns.barplot(data  = data\
              , x = "job_title", y = "jobs")
    plt.title(f"Top 5 popular jobs for  year: {year}")
    plt.xticks(rotation = 45)
    plt.show()
```

CONCLUSION:As we can see by our observation that Data Scientist have more scope in the Year 2021,2022,2023 so overall Pursuing a career as data scientist requires a combination of technical skills ,analytical skills and communication skills.

## **5. Company location on basis of Size**

## **6. Top 5 job title based on company size and emp_type for freshers**

```python
temp_df = df[["job_title", "company_size","employment_type"]][df["experience_level"]=="EN"].value_counts().reset_index(name = "count")

try:
    for size in temp_df["company_size"].unique():
        for tp in temp_df["employment_type"].unique():
            data = temp_df[(temp_df["company_size"] == size) & (temp_df["employment_type"] == tp)].head()

            sns.barplot(data = data\
                  , x = "job_title", y ="count")
            plt.title(f"Top 5 popular jobs for {size}, {tp}")
            plt.xticks(rotation = 45)
            plt.show()
except:
    print()
```

CONCLUSION:
           We can see that Machine Learning Engineer and Data Analyst are the two leading job titles for data people with Data            Engineer becoming more popular as of late most probably because more and more companies are realizing that they                need a solid data infrastructure to add value to their business before hiring a Data Scientist or a Data Analyst              and Machine Learning Engineer.
